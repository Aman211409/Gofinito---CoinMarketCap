Importing modules and packages

```python
import requests
import pandas as pd
import sqlalchemy
```

Calling the API using requests

```python
result = requests.get('https://web-api.coinmarketcap.com/v1.1/cryptocurrency/quotes/historical?convert=USD,BTC&format=chart_crypto_details&interval=1h&time_end=1598092056&time_start=1595413656&id=1')
```

```python
result.status_code
```

As the status_code is 200, therefore, we can assume that there are no errors

```python
result.text
```

```python
result.json()
```

Storing the json data 

```python
data = result.json()
```

As the first key in the json file is 'status', we will filter that out and only use the key 'data'

```python
data = data['data']
```

Let's store the data in a pandas dataframe

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

As the format of the data is unwanted, we will use the transpose of the dataframe

```python
df1 = df.T
```

```python
df1.head()
```

As we want the data for the current day and not the previous data, we will get the last 100 rows of the dataframe.

```python
df2 = df1.tail(100)
```

Now that we have the data, we will establish a connection using sqlalchemy. 

```python
engine = sqlalchemy.create_engine('postgresql://postgres:211409@localhost/learning')  # Specifying the details of the PostgreSQL server

con = engine.connect()  # Establishing the connection

print(engine.table_names())  # Printing out the already existing tables in the database.
```

We can see that the table doesn't already exist. Let's define a table name that will be used in PostgreSQL

```python
table_name = 'CoinMarketCap'
```

```python
df2.to_sql(table_name, con, if_exists='replace', index=True)  # We will replace the pre-existing table with a new 
                                                              # to get the data for the current day and not the previous.
    
print(engine.table_names())  # Printing the final list of tables in the database
```

We can see that the table has been successfully added to PostgreSQL database. 

Now, we will close the connection.

```python
con.close()
```
